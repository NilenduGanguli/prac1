```python
# 1.Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2. Load Data
df = pd.read_csv('https://s3.eu-west-1.amazonaws.com/neueda.conygre.com/pydata/SPX.csv',index_col='Date', parse_dates=True)
```

```python
df.head(10)
```

```python
# 3. Convert Data to numericals
df['Price'] = pd.to_numeric(df['Price'].str.replace(',',''))
df['Open'] = pd.to_numeric(df['Open'].str.replace(',',''))
df['High'] = pd.to_numeric(df['High'].str.replace(',',''))
df['Low'] = pd.to_numeric(df['Low'].str.replace(',',''))
df['Change %'] = pd.to_numeric(df['Change %'].str.replace('%',''))
```

```python
df.head(10)
```

```python
# 4. Sort data
df.sort_values(by='Date',inplace=True)
```

```python
# 5. Plot line chart June 2015 to July 2016
df1=df.query(" Date >= '2015-06-01' and Date <='2015-07-30'")[['High','Low']]
df1.plot()
```

```python
# 6. Find min, max, average of all columns
byMonth=pd.Grouper(freq='BM')
funcs=['min','max','mean']
df.groupby(byMonth).agg(funcs)
```
